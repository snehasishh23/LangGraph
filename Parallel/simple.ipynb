```python
from langgraph.graph import StateGraph, START, END
from typing import TypedDict
```

```python
class cricket_state(TypedDict):
    runs: int
    bowls: int
    fours: int
    sixes: int
    strike_rate: float
    boundary_percentage: float
    bowls_per_boundary: float
    summary: str
```

```python
graph= StateGraph(cricket_state)
```

```python
def strike_rate(state:cricket_state):
    runs= state['runs']
    bowls= state['bowls']
    if bowls==0:
        strike_rate=0.0
    else:
        strike_rate=(runs/ bowls)*100
    return {'strike_rate': strike_rate}
```

```python
def boundary_percentage(state:cricket_state):
    runs= state['runs']
    fours= state['fours']
    sixes= state['sixes']
    total_boundaries= (fours*4) + (sixes*6)
    if runs==0:
        boundary_percentage=0.0
    else:
        boundary_percentage=(total_boundaries/ runs)*100
    return {'boundary_percentage': boundary_percentage} 
```

```python
def bowls_per_boundary(state:cricket_state):
    bowls= state['bowls']
    fours= state['fours']
    sixes= state['sixes']
    total_boundaries= fours + sixes
    if total_boundaries==0:
        bowls_per_boundary=0.0
    else:
        bowls_per_boundary= bowls/ total_boundaries
    return {'bowls_per_boundary': bowls_per_boundary}
```

```python
def summary(state:cricket_state):
    runs= state['runs']
    bowls= state['bowls']
    fours= state['fours']
    sixes= state['sixes']
    strike_rate= state['strike_rate']
    boundary_percentage= state['boundary_percentage']
    bowls_per_boundary= state['bowls_per_boundary']
    
    summary_str=f"Runs: {runs}, Bowls: {bowls}, Fours: {fours}, Sixes: {sixes}, Strike Rate: {strike_rate:.2f}%, Boundary Percentage: {boundary_percentage:.2f}%, Bowls per Boundary: {bowls_per_boundary:.2f}"
    
    summary= summary_str
    return {'summary': summary}
```

```python
graph.add_node('strike_rate', strike_rate)
graph.add_node('boundary_percentage', boundary_percentage)
graph.add_node('bowls_per_boundary', bowls_per_boundary)
graph.add_node('summary', summary)
```

```python
graph.add_edge(START, 'strike_rate')
graph.add_edge(START, 'boundary_percentage')
graph.add_edge(START, 'bowls_per_boundary')
graph.add_edge('strike_rate', 'summary')
graph.add_edge('boundary_percentage', 'summary')
graph.add_edge('bowls_per_boundary', 'summary')
graph.add_edge('summary', END)
```

```python
workflow=graph.compile()
```

```python
initial_state= cricket_state(runs=50, bowls=30, fours=5, sixes=2)
final_state= workflow.invoke(initial_state)
print(final_state['summary'])
```
