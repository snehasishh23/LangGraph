```python
from langgraph.graph import StateGraph,START,END
from typing import TypedDict,Literal
```

```python
class quad(TypedDict):
    a: int
    b: int
    c: int
    Eqn : str
    discriminant: int
    result: str
```

```python
def eqn(state:quad):
    Eqn = f'{state["a"]}x^2 + {state["b"]}x + {state["c"]} = 0'
    return {'Eqn': Eqn}
```

```python
def discriminant(state:quad):
    discriminant = state["b"]**2 - 4*state["a"]*state["c"]
    return {'discriminant': discriminant}
```

```python
def real_roots(state:quad):
    root1= ((-state["b"])+ (state["discriminant"]**0.5))/(2*state["a"])
    root2= ((-state["b"])- (state["discriminant"]**0.5))/(2*state["a"])
    result= f'The roots are {root1} and {root2}'
    return {'result': result}
```

```python
def equal_roots(state:quad):
    root= -state["b"]/(2*state["a"])
    result= f'only repeating root  is {root}'
    return {'result': result}
```

```python
def imaginary_roots(state:quad):
    result= f'no real roots exist as the discriminant is negative'
    return {'result': result}
```

```python
def check_condition(state: quad) -> Literal["real_roots", "equal_roots", "imaginary_roots"]:
    if state['discriminant'] > 0:
        return "real_roots"
    elif state['discriminant'] == 0:
        return "equal_roots"
    else:
        return "imaginary_roots"
```

```python
graph= StateGraph(quad)

#nodes
graph.add_node('eqn',eqn)
graph.add_node('discriminant',discriminant)
graph.add_node('real_roots',real_roots)
graph.add_node('imaginary_roots',imaginary_roots)
graph.add_node('equal_roots',equal_roots)

#edges
graph.add_edge(START,'eqn')
graph.add_edge('eqn','discriminant')
graph.add_conditional_edges('discriminant', check_condition)
graph.add_edge('real_roots', END)
graph.add_edge('equal_roots', END)
graph.add_edge('imaginary_roots', END)

workflow = graph.compile()
```
